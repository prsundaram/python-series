# Question Answering

```python
from transformers import pipeline
import pandas as pd
```

```python
context = """
Prasanna Sundaram is a software engineer working at Infosys in Chennai.
He joined the company in 2021 and works on artificial intelligence projects.
"""

questions = [
    "Where does Prasanna Sundaram work?",
    "Where is he located?",
    "When did Prasanna join Infosys?",
    "What does he work on?",
    "What is his profession?"
]
```

```python
qa = pipeline("question-answering")
```

```python
for q in questions:
    result = qa(question=q, context=context)
    print(f"Q: {q}")
    print(f"A: {result['answer']} (score: {result['score']:.2f})\n")
```

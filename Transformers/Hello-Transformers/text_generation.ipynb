# Text Generation

```python
from transformers import pipeline, set_seed
import pandas as pd
```

```python
set_seed(0)
```

```python
generator = pipeline("text-generation", model="gpt2-medium")

prompt = "Write a brief note about Donald J Trump"

output = generator(
    prompt,
    max_new_tokens=80,
    do_sample=True,
    temperature=0.7,
    top_p=0.9,
    repetition_penalty=1.2
)

print(output[0]["generated_text"])
```

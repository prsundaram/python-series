# Image Segmentation

```python
from transformers import pipeline

segmenter = pipeline("image-segmentation")

url = "http://images.cocodataset.org/val2017/000000252219.jpg"

output = segmenter(url)

print(len(output))  # number of segments
```

```python
import matplotlib.pyplot as plt

for i, seg in enumerate(output):
    plt.imshow(seg['mask'])
    plt.title(seg['label'])
    plt.axis('off')
    plt.show()
```

# Image Segmentation

```python
from transformers import pipeline

segmenter = pipeline("image-segmentation")

url = "https://images.unsplash.com/photo-1503023345310-bd7c1de61c7d"

output = segmenter(url)

print(len(output))  # number of segments
```

```python
import matplotlib.pyplot as plt

for i, seg in enumerate(output):
    plt.imshow(seg['mask'])
    plt.title(seg['label'])
    plt.axis('off')
    plt.show()
```

## Text Classification


```python
from transformers import pipeline
import pandas as pd
```

```python
## Default model
# https://huggingface.co/distilbert/distilbert-base-uncased-finetuned-sst-2-english
classifier = pipeline("text-classification", model="distilbert-base-uncased-finetuned-sst-2-english")
```

```python
input_text_list = [

    'I Love You',

    'I hate you',

    'I do not think you get it probably',

    'Well, will circle back on this'

]

outputs = classifier(input_text_list)

output_sentiment = pd.DataFrame(outputs)

output_sentiment["text"] = input_text_list

display(output_sentiment)
```

```python
# Load model

classifier = pipeline(

    "text-classification",

    model="SamLowe/roberta-base-go_emotions",

    top_k=1

)

# Emotion → Emoji mapping
emotion_emoji = {
    "admiration": "🤝",
    "amusement": "🎉",
    "anger": "😡",
    "annoyance": "😠",
    "approval": "👍",
    "caring": "❤️",
    "confusion": "🤔",
    "curiosity": "🤨",
    "desire": "🤤",
    "disappointment": "😞",
    "disapproval": "🙅",
    "disgust": "🤢",
    "embarrassment": "😳",
    "excitement": "🤩",
    "fear": "😨",
    "gratitude": "🙏",
    "grief": "😔",
    "joy": "😊",
    "love": "😍",
    "nervousness": "😬",
    "optimism": "🌟",
    "pride": "💪",
    "realization": "🧠",
    "relief": "😤",
    "remorse": "😢",
    "sadness": "😢",
    "surprise": "😲",
    "neutral": "😐"
}

# Test sentences

sentences = [

    "I am furious about how unfairly I was treated.",

    "That smell is absolutely disgusting and unbearable.",

    "I’m really scared something bad is about to happen.",

    "I feel so happy and full of life today!",

    "I feel empty and heartbroken after what happened.",

    "Wow, I did not expect that at all!",

    "I deeply care about you and cherish every moment with you.",

    "I truly admire your dedication and hard work.",

    "That joke was hilarious, I can’t stop laughing!",

    "This constant noise is really getting on my nerves.",

    "That’s a great idea, I completely support it.",

    "I don’t understand what you’re trying to say.",

    "I’m really curious about how this works.",

    "I really want that new phone so badly.",

    "I’m disappointed that things didn’t go as planned.",

    "I don’t agree with this decision at all.",

    "I felt so embarrassed when I tripped in front of everyone.",

    "Thank you so much, I truly appreciate your help.",

    "Losing them has left a deep pain in my heart.",

    "I’m so relieved that everything turned out okay.",

    "I regret what I said and feel really sorry.",

    "I’m proud of myself for achieving this goal.",

    "I know I can handle this challenge successfully.",

    "I feel nervous before speaking in front of the crowd.",

    "Oh, now I finally understand what you meant!",

    "The view from the mountain was absolutely breathtaking.",

    "I went to the store and bought some groceries."

]

# Run batch inference

outputs = classifier(sentences)

# Print results

for text, result in zip(sentences, outputs):

    label = result[0]["label"]

    score = result[0]["score"]

    emoji = emotion_emoji.get(label, "")

    print(f"{emoji} {label.upper()} ({score:.2f}) → {text}")
```

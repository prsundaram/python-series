# Summarization

```python
from transformers import pipeline
import pandas as pd
```

```python
text = """
Prasanna Sundaram is a software engineer working at Infosys in Chennai. 
He specializes in artificial intelligence and machine learning. 
Over the years, he has contributed to multiple projects involving natural language processing and automation. 
He is passionate about learning new technologies and applying them to solve real-world problems.
"""
```

```python
summarizer = pipeline("summarization")
```

```python
summarization = summarizer(text)

# 🔥 Step 1 — calculate input length (tokens or words)
input_length = len(text.split())

# 🔥 Step 2 — dynamic min/max
max_len = int(input_length * 0.6)
min_len = int(input_length * 0.3)

# safety bounds
max_len = max(30, max_len)
min_len = max(10, min_len)
print(f"Input length: {input_length}")
print(f"Min length: {min_len}, Max length: {max_len}")

# 🔥 Step 3 — summarization
summary = summarizer(
    text,
    max_length=max_len,
    min_length=min_len,
    do_sample=False
)

print("\nSummary:")
print(summary[0]["summary_text"])
```

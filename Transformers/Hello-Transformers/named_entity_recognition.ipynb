# Named Entity Recognition

```python
from transformers import pipeline
import pandas as pd
```

```python
ner_tagger = pipeline("ner", aggregation_strategy="simple") 
```

```python
text = """
I am Prasanna Sundaram, a software engineer at Infosys in Chennai, India.
On 15th August 2023 at 10:30 AM, I attended the Google I/O event.
I bought an iPhone 15 for $1200 with a 10% discount.
This is my 1st purchase and I speak English.
I also read the book 'The Alchemist' and studied the Indian Constitution.
"""
output = ner_tagger(text)
pd.DataFrame(output)
```

```python
print("🔍 NER CLI (type 'exit' to quit)\n")

while True:

    text = input("Enter text: ")

    if text.lower() == "exit":

        print("👋 Exiting...")

        break

    output = ner_tagger(text)

    if not output:

        print("No entities found.\n")

        continue

    df = pd.DataFrame(output)

    print("\nEntities detected:")

    print(df[["word", "entity_group", "score"]])

    print("\n" + "-"*50 + "\n")
```

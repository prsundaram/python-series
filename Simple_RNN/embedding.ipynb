# Word Embeddings using Keras

This notebook demonstrates how text is converted into numerical form and then into dense vector representations (embeddings).

Pipeline:

$$
\text{Text} \rightarrow \text{One Hot Encoding} \rightarrow \text{Padding} \rightarrow \text{Embedding Layer}
$$

```python
from tensorflow.keras.preprocessing.text import one_hot
```

## Step 1: Define Sentences

We start with a small corpus of sentences.

Let the dataset be:

$$
S = \{s_1, s_2, ..., s_n\}
$$

```python
## sentences
sent = [
    'the glass of milk',
    'the glass of juice',
    'the cup of tea',
    'I am a good boy',
    'I am a good developer',
    'Understand the meaning of words',
    'your videos are good'
]
```

```python
sent
```

## Step 2: Vocabulary Size

We define the vocabulary size for hashing.

Each word is mapped to an integer:

$$
\text{word} \rightarrow \text{index} \in [0, V)
$$

Where $V$ is the vocabulary size.

```python
## Define the vocabulary size
voc_size = 10000
```

## Step 3: One Hot Representation

Each sentence is converted into a list of integers using hashing:

$$
\text{Sentence} \rightarrow [w_1, w_2, ..., w_k]
$$

Note: This is **not true one-hot encoding**, but a hashed representation.

```python
### One Hot Representation
one_hot_rep = [one_hot(words, voc_size) for words in sent]
print(one_hot_rep)
```

## Step 4: Padding Sequences

Neural networks require inputs of equal length.

We pad sequences to a fixed length:

$$
\text{length} = L
$$

Shorter sequences are padded with zeros.

```python
## Word Embedding Representation
from tensorflow.keras.layers import Embedding
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
sent_length = 8

embedded_docs = pad_sequences(
    one_hot_rep,
    padding='pre',
    maxlen=sent_length)

print(embedded_docs)
```

## Step 5: Embedding Layer

The embedding layer maps each word index to a dense vector:

$$
\text{Embedding}(i) = \vec{v}_i \in \mathbb{R}^d
$$

Where:
- $i$ → word index  
- $d$ → embedding dimension

```python
## Feature representation
dim = 10
```

```python
model = Sequential()
model.add(
    Embedding(
        voc_size,
        dim
    )
)

model.compile(
    'adam',
    'mse'
)
```

## Model Summary

The embedding matrix has shape:

$$
(V, d)
$$

Where:
- $V$ = vocabulary size  
- $d$ = embedding dimension

```python
model.summary()
```

## Step 6: Generate Embeddings

Each input sequence is transformed into:

$$
(L, d)
$$

Where:
- $L$ = sentence length  
- $d$ = embedding dimension

```python
model.predict(embedded_docs)
```

## Inspect a Single Input

We check one padded sentence:

$$
\text{Input} \rightarrow [x_1, x_2, ..., x_L]
$$

```python
embedded_docs[0]
```

## Embedding for Single Sentence

The output becomes:

$$
(L, d) = (8, 10)
$$

Each word is now represented as a dense vector.

```python
model.predict(np.expand_dims(embedded_docs[0], axis=0))
```

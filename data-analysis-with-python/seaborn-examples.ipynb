# üìä Basic Plotting with Seaborn

üìå **Objective:**  
Learn how to create simple and insightful visualizations using Seaborn ‚Äî starting with a **scatter plot** to show relationships between two continuous variables.

---

## üß† About the Dataset

We‚Äôll use Seaborn‚Äôs built-in **`tips`** dataset ‚Äî  
a collection of restaurant bills and tips, with columns like:
- `total_bill` ‚Äî total amount of the bill  
- `tip` ‚Äî amount of tip given  
- `sex` ‚Äî gender of the person who paid  
- `smoker` ‚Äî whether they were smokers  
- `day` ‚Äî day of the week  
- `time` ‚Äî Lunch/Dinner  
- `size` ‚Äî number of people at the table  

---

## üîπ Step-by-Step Explanation

| Step | Description |
|:------|:-------------|
| 1Ô∏è‚É£ | Load Seaborn‚Äôs sample dataset `tips` |
| 2Ô∏è‚É£ | Use `sns.scatterplot()` to visualize relationship between `total_bill` and `tip` |
| 3Ô∏è‚É£ | Customize plot with title, labels, and grid |
| 4Ô∏è‚É£ | Display plot using `plt.show()` |


````python
# --------------------------------------------------------
# üìä Basic Scatter Plot using Seaborn
# --------------------------------------------------------

# üß† Step 1: Load sample dataset
tips = sns.load_dataset('tips')  # Preloaded dataset of restaurant bills and tips

# üßæ Step 2: Inspect first few records
display(tips.head())

# üß© Step 3: Create scatter plot
plt.figure(figsize=(8, 5))  # Optional: set figure size
sns.scatterplot(
    x='total_bill',          # X-axis: total bill amount
    y='tip',                 # Y-axis: tip amount
    data=tips,               # Data source (DataFrame)
    color='royalblue',       # Point color
    edgecolor='black',       # Outline for better contrast
    s=80,                    # Dot size
    alpha=0.7                # Transparency for overlapping points
)

# üè∑Ô∏è Step 4: Add chart title and labels
plt.title('Scatter Plot of Total Bill vs Tip', fontsize=14, fontweight='bold')
plt.xlabel('Total Bill (in $)', fontsize=12)
plt.ylabel('Tip Amount (in $)', fontsize=12)

# üßæ Step 5: Add gridlines for readability
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
````

‚úÖ Output Explanation
- Each point represents one restaurant bill.
- The X-axis shows the total bill, and the Y-axis shows the tip amount.
- You can visually observe that larger bills generally lead to higher tips ‚Äî a clear positive correlation.

## üß© Example: Add Grouping by Day (Colored Scatter)

````python
plt.figure(figsize=(8, 5))
sns.scatterplot(
    x='total_bill', y='tip',
    hue='day', data=tips, s=90, alpha=0.8, edgecolor='black'
)

plt.title('Scatter Plot of Total Bill vs Tip (by Day)', fontsize=14, fontweight='bold')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')
plt.legend(title='Day', loc='upper left')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
````

# üìà Line Plot using Seaborn

üìå **Objective:**  
Learn how to create a **line plot** in Seaborn to show how a numerical variable changes with respect to another variable ‚Äî  
for example, how the **total bill** varies based on the **party size** at a restaurant.

---

## üß† When to Use a Line Plot

| Use Case | Description |
|:-----------|:-------------|
| ‚è±Ô∏è Time Series | Show trends over time (e.g., monthly revenue, weekly traffic) |
| üìè Continuous Data | Visualize changes in one variable relative to another |
| üß© Averages per Group | Display average values for grouped data |
| üìä Trend Visualization | Identify upward, downward, or stable trends |

---

## üîπ Step-by-Step Explanation

| Step | Description |
|:------|:-------------|
| 1Ô∏è‚É£ | Load the dataset (`tips`) |
| 2Ô∏è‚É£ | Use `sns.lineplot()` to plot `total_bill` vs `size` |
| 3Ô∏è‚É£ | Add title, labels, and customize style |
| 4Ô∏è‚É£ | Display the final chart |


````python
# --------------------------------------------------------
# üìà Basic Line Plot using Seaborn
# --------------------------------------------------------

# üß† Step 1: Load the sample dataset
tips = sns.load_dataset('tips')

# üßæ Step 2: Inspect first few rows (optional)
display(tips.head())

# üß© Step 3: Create a simple line plot
plt.figure(figsize=(8, 5))
sns.lineplot(
    x='size',              # X-axis: party size (number of people)
    y='total_bill',        # Y-axis: total bill amount
    data=tips,             # Data source
    color='teal',          # Line color
    marker='o',            # Add circular markers at data points
    linewidth=2,           # Line thickness
    markersize=8,          # Marker size
    alpha=0.9              # Transparency level
)

# üè∑Ô∏è Step 4: Add labels and title
plt.title('Line Plot of Total Bill by Party Size', fontsize=14, fontweight='bold')
plt.xlabel('Party Size', fontsize=12)
plt.ylabel('Total Bill ($)', fontsize=12)

# üßæ Step 5: Add gridlines for better readability
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
````

‚úÖ Output Explanation

You‚Äôll see:
- The X-axis shows the number of people at the table (size).
- The Y-axis shows the average total bill amount.
- The teal line indicates that as group size increases, the total bill also increases ‚Äî a logical trend!

---

| Tip                          | Description                                                  |
| :--------------------------- | :----------------------------------------------------------- |
| üßÆ Aggregate automatically   | Use `estimator='mean'` to plot averages per group            |
| üåà Add multiple lines        | Use `hue='sex'` or `hue='day'` to compare trends             |
| üß± Show confidence intervals | Set `ci='sd'` or `ci=None` to control shading                |
| üé® Change theme              | Use `sns.set_style('darkgrid')` or `sns.set_context('talk')` |
| üíæ Save figure               | `plt.savefig('lineplot_totalbill_size.png', dpi=300)`        |
---
## üß© Example: Comparing by Day (Multiple Lines)

````python
# Multiple line plot comparing Total Bill by Day and Size
plt.figure(figsize=(10, 6))
sns.lineplot(
    x='size', y='total_bill',
    hue='day',                # Separate lines for each day
    data=tips,
    marker='o',
    linewidth=2,
    palette='Set2'            # Beautiful built-in color palette
)

plt.title('Total Bill vs Party Size by Day', fontsize=14, fontweight='bold')
plt.xlabel('Party Size')
plt.ylabel('Total Bill ($)')
plt.legend(title='Day of Week')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
````

# üìä Categorical Variable Plot ‚Äî Bar Plot in Seaborn

üìå **Objective:**  
Visualize how a **numerical variable** (like total bill) changes across different **categories** (like days of the week).  
This helps identify which categories perform better or worse ‚Äî for instance, which day earns the most at a restaurant.

---

## üß† What is a Bar Plot?

A **Bar Plot** shows the **mean (average)** value of a numerical variable for each category.  
It automatically calculates the **central tendency** (by default, the *mean*) for each group and displays confidence intervals as error bars.

$$[
\text{Bar height} = \text{Mean of } Y \text{ for each category in } X
]$$

---

### üîπ When to Use a Bar Plot

| Use Case | Description |
|:-----------|:-------------|
| üçΩÔ∏è Restaurant Analysis | Compare average bills or tips by day |
| üõçÔ∏è Sales Reports | Compare revenue by region or category |
| üéì Education | Average scores by department or grade |
| üè¢ HR | Average salary by job title or department |

---

## üß© Example 1: Basic Bar Plot


````python
# --------------------------------------------------------
# üìä Bar Plot - Average Total Bill by Day
# --------------------------------------------------------

# üß† Step 1: Load sample dataset
tips = sns.load_dataset('tips')

# üßæ Step 2: Inspect first few records (optional)
display(tips.head())

# üß© Step 3: Create basic bar plot
plt.figure(figsize=(8, 5))
sns.barplot(
    x='day',                  # X-axis: categorical variable (Day of the week)
    y='total_bill',           # Y-axis: numerical variable (Total bill amount)
    data=tips,                # Data source
    color='teal',             # Bar color (optional override)
    edgecolor='black',        # Outline color
    errorbar='sd',            # Show standard deviation as error 
    err_kws={'color': 'red'}  # color for the error
)

# üè∑Ô∏è Step 4: Add labels, title, and grid
plt.title('Bar Plot of Average Total Bill by Day', fontsize=14, fontweight='bold')
plt.xlabel('Day of the Week', fontsize=12)
plt.ylabel('Average Total Bill ($)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
````

‚úÖ Output Explanation
- Each bar shows the average total bill for that day of the week.
- The black lines (error bars) indicate variation (standard deviation) within each day.
- This helps quickly identify which days have higher average bills (e.g., weekends ü•Ç).

---
## üß© Example 2: Bar Plot Grouped by Gender (Using hue)

````python
# --------------------------------------------------------
# üìä Grouped Bar Plot - Total Bill by Day and Gender
# --------------------------------------------------------

plt.figure(figsize=(8, 5))
sns.barplot(
    x='day', y='total_bill',
    hue='sex',               # Separate bars for Male/Female
    data=tips,
    palette='Set2',          # Soothing color palette
    edgecolor='black',
    errorbar='sd'
)

plt.title('Average Total Bill by Day and Gender', fontsize=14, fontweight='bold')
plt.xlabel('Day of the Week', fontsize=12)
plt.ylabel('Average Total Bill ($)', fontsize=12)
plt.legend(title='Gender')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
````

# üì¶ Box Plot using Seaborn

üìå **Objective:**  
Learn how to visualize the **distribution and spread** of numerical data across categories using Seaborn‚Äôs `boxplot()`.  
Box plots show where most of your data lies and help identify **outliers** (unusually high or low values).

---

## üß† What is a Box Plot?

A **Box Plot** displays the **five-number summary** of a dataset:

| Metric | Description |
|:--------|:-------------|
| Minimum | The smallest data point excluding outliers |
| Q1 (25th percentile) | The lower quartile ‚Äî 25% of data lies below this value |
| Median (Q2) | The middle value of the dataset |
| Q3 (75th percentile) | The upper quartile ‚Äî 75% of data lies below this value |
| Maximum | The largest data point excluding outliers |

---

### üîπ Understanding the Structure

A box plot visually shows:



```sccs
      ‚óè    ‚Üê Outliers (extreme values)
‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚î§‚îÇ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ  ‚Üê Whiskers (min & max)
     ‚îÇ‚îÇ
  ‚îå‚îÄ‚îÄ‚îº‚îº‚îÄ‚îÄ‚îê    ‚Üê Box (middle 50% of data: Q1‚ÄìQ3)
     ‚îî‚îº‚îò
      ‚îÇ
     Median
```


- The **box** shows the middle 50% of data (interquartile range ‚Äî IQR).  
- The **line inside** the box is the **median (Q2)**.  
- The **whiskers** extend to the lowest and highest non-outlier values.  
- The **dots beyond whiskers** are **outliers** ‚Äî unusual points that deviate significantly from others.  

---

## ‚ö†Ô∏è What Are Outliers?

**Outliers** are data points that fall significantly outside the normal range of values.  

Mathematically, any value beyond:

$$[
\text{Outlier if } x < Q1 - 1.5 \times IQR \quad \text{or} \quad x > Q3 + 1.5 \times IQR
]$$

where  
$$[
IQR = Q3 - Q1
]$$

Outliers are not always ‚Äúbad‚Äù ‚Äî they can indicate:
- Errors or anomalies in data entry  
- Special events (e.g., huge tips or promotions)  
- Legitimate rare occurrences  

---

## üß© Example 1: Basic Box Plot


````python
# --------------------------------------------------------
# üì¶ Basic Box Plot using Seaborn
# --------------------------------------------------------


# üß† Step 1: Load sample dataset
tips = sns.load_dataset('tips')

# üß© Step 2: Create box plot for total bill by day
plt.figure(figsize=(8, 5))
sns.boxplot(
    x='day',                # Categorical variable (X-axis)
    y='total_bill',         # Numerical variable (Y-axis)
    data=tips,              # Data source
    palette='Set2',         # Beautiful color palette
    width=0.6,              # Box width
    fliersize=6,            # Size of outlier markers
    linewidth=1.5,          # Line thickness
    hue = 'day'
)

# üè∑Ô∏è Step 3: Add labels, title, and grid
plt.title('Box Plot of Total Bill by Day', fontsize=14, fontweight='bold')
plt.xlabel('Day of the Week', fontsize=12)
plt.ylabel('Total Bill ($)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
````

‚úÖ Output Explanation
- Each box represents the distribution of total bills for that day.
- The median line shows the central value (typical bill).
- The height of the box represents variability (IQR).
- Dots outside whiskers are outliers, i.e., unusually high or low bills ‚Äî maybe a party of 10 üçæ

## üß© Example 2: Box Plot Grouped by Gender

````python
# --------------------------------------------------------
# üì¶ Box Plot - Total Bill by Day and Gender
# --------------------------------------------------------

plt.figure(figsize=(8, 5))
sns.boxplot(
    x='day', y='total_bill',
    hue='sex',                # Separate boxes by gender
    data=tips,
    palette='pastel',
    width=0.6,
    linewidth=1.4,
    fliersize=5
)

plt.title('Total Bill Distribution by Day and Gender', fontsize=14, fontweight='bold')
plt.xlabel('Day of the Week')
plt.ylabel('Total Bill ($)')
plt.legend(title='Gender')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
````

‚úÖ Enhanced Output Insight
- Now each day has two boxes ‚Äî one for each gender.
- You can visually compare spending behavior between men and women per day.
- Outliers (points beyond whiskers) highlight days with exceptionally high bills.

## üßÆ Understanding IQR and Outliers

üìå **Objective:**  
Before we interpret box plots, it‚Äôs important to understand two key ideas ‚Äî  
**IQR (Interquartile Range)** and **Outliers**.

---

### üîπ What is IQR?

**IQR (Interquartile Range)** tells us how *spread out* the middle 50% of data is.

It is calculated as:

$[
\text{IQR} = Q_3 - Q_1
]$

where:
- $( Q_1 )$ ‚Üí 25th percentile (lower quartile)  
- $( Q_3 )$ ‚Üí 75th percentile (upper quartile)

üí° Think of it like this:
> The IQR covers the ‚Äúnormal range‚Äù where most data points live ‚Äî it‚Äôs your *comfort zone* for values.

---

### üîπ What are Outliers?

**Outliers** are data points that are **much higher or lower** than the rest of the data.  
They fall *outside* the typical IQR range.

Mathematically, any value is considered an **outlier** if it lies:

$[
x < Q_1 - 1.5 \times \text{IQR} \quad \text{or} \quad x > Q_3 + 1.5 \times \text{IQR}
]$

üü¢ Most values lie within the box (inside the IQR).  
üî¥ Outliers appear as individual dots **beyond the whiskers** in a box plot.

---

### ‚úÖ **Simple Summary**

| Term | Meaning | Example |
|:------|:----------|:-------------|
| **Q1 (25%)** | Lower boundary of the middle range | ‚Çπ25 if 25% of bills are below this |
| **Q3 (75%)** | Upper boundary of the middle range | ‚Çπ75 if 75% of bills are below this |
| **IQR (Q3‚ÄìQ1)** | Middle spread of values | ‚Çπ75 ‚Äì ‚Çπ25 = ‚Çπ50 |
| **Outliers** | Unusual values far away from the rest | ‚Çπ200 when most bills are ‚Çπ30‚Äì‚Çπ80 |

---

üí¨ **In simple words:**  
üëâ *IQR shows the middle chunk of data.*  
üëâ *Outliers are data points that ‚Äúdon‚Äôt fit in‚Äù ‚Äî they‚Äôre too far away from the rest.*

---

üß† Tip: Outliers aren‚Äôt always errors ‚Äî they can represent **interesting cases**, like a huge restaurant bill or an exceptional sales day!


# üéª Violin Plot in Seaborn

üìå **Objective:**  
Learn how to visualize the **distribution and density** of numerical data across categories using the `sns.violinplot()` function.  
A **violin plot** combines the information of a **box plot** (median, quartiles, outliers) with a **smoothed density curve**, making it easier to understand *how values are spread*.

---

## üß† What is a Violin Plot?

A **Violin Plot** shows:
- The **median** (white dot or line inside)
- The **interquartile range (IQR)** ‚Äî middle 50% of values
- The **distribution density** ‚Äî the ‚Äúwidth‚Äù of the violin shows where data is more frequent

It‚Äôs called a *violin* because of its **symmetrical shape**, where the thickness represents the **concentration of data points**.

---

### üîπ Box Plot vs. Violin Plot

| Feature | Box Plot | Violin Plot |
|:----------|:----------|:------------|
| Shape | Rectangular box | Symmetrical ‚Äúviolin‚Äù |
| Shows median & IQR | ‚úÖ Yes | ‚úÖ Yes |
| Shows data density | ‚ùå No | ‚úÖ Yes |
| Handles multiple groups | ‚úÖ Yes | ‚úÖ Yes |

üí° **In short:**  
> The *violin plot* tells you *where* your data is concentrated ‚Äî not just its range.

---

## üß© Example 1: Basic Violin Plot


````python
# --------------------------------------------------------
# üéª Basic Violin Plot using Seaborn
# --------------------------------------------------------

# üß† Step 1: Load the sample dataset
tips = sns.load_dataset('tips')

# üßæ Step 2: Inspect dataset
display(tips.head())

# üß© Step 3: Create violin plot for total bill by day
plt.figure(figsize=(8, 5))
sns.violinplot(
    x='day',                   # Categorical variable
    y='total_bill',            # Numerical variable
    data=tips,                 # Data source
    palette='muted',           # Color theme
    inner='box',               # Show mini box plot inside violin
    linewidth=1.2,             # Edge thickness
    hue='day'
)

# üè∑Ô∏è Step 4: Add title and labels
plt.title('Violin Plot of Total Bill by Day', fontsize=14, fontweight='bold')
plt.xlabel('Day of the Week', fontsize=12)
plt.ylabel('Total Bill ($)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
````

‚úÖ Output Explanation
- Each ‚Äúviolin‚Äù represents the distribution of total bills for that day.
- The wider sections indicate where more data points cluster.
- The thin sections show where data is sparse.
- The white dot (or line) inside shows the median.
- The box inside the violin shows the IQR ‚Äî the middle 50% of the data.

---
## üß© Example 2: Violin Plot Grouped by Gender

````python
# --------------------------------------------------------
# üéª Grouped Violin Plot by Gender
# --------------------------------------------------------

plt.figure(figsize=(8, 5))
sns.violinplot(
    x='day', y='total_bill',
    hue='sex',                 # Split by gender
    data=tips,
    split=True,                # Show both genders in one violin
    palette='Set2',            # Color palette
    inner='quartile'           # Show quartile lines inside
)
sns.set_style('whitegrid')
plt.title('Total Bill Distribution by Day and Gender', fontsize=14, fontweight='bold')
plt.xlabel('Day of the Week')
plt.ylabel('Total Bill ($)')
plt.legend(title='Gender')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
````

‚úÖ Enhanced Output Insight
- Each ‚Äúhalf‚Äù of a violin shows the distribution for one gender.
- You can easily compare spending patterns between men and women across days.
- Quartile lines inside show median and spread ‚Äî a mix of stats and density üí™

---
| Tip                  | Description                                            |
| :------------------- | :----------------------------------------------------- |
| üé® Change style      | Use `sns.set_style('whitegrid')` for readability       |
| üßÆ Remove box inside | Set `inner=None` to show only the violin shape         |
| ‚öôÔ∏è Hide outliers     | Set `cut=0` to trim long tails                         |
| üåà Color control     | Use `palette='coolwarm'`, `'mako'`, or `'pastel'`      |
| üíæ Save figure       | `plt.savefig('violinplot_totalbill_day.png', dpi=300)` |

---
## üß© Example 3: Vertical ‚Üí Horizontal Violin Plot

````python
# --------------------------------------------------------
# üéª Horizontal Violin Plot
# --------------------------------------------------------

plt.figure(figsize=(7, 5))
sns.violinplot(
    y='day', x='total_bill',
    data=tips,
    palette='crest',
    inner='point',
    linewidth=1.3,
    hue='day'
)

plt.title('Horizontal Violin Plot - Total Bill by Day', fontsize=14, fontweight='bold')
plt.xlabel('Total Bill ($)')
plt.ylabel('Day of the Week')
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
````

‚úÖ Output Explanation
- The violin is now horizontal, great for dashboards or reports.
- The width still represents data density, and quartile markers stay visible.

üßÆ Quick Summary
$$[
\textbf{Violin Plot} = \textbf{Box Plot} + \textbf{Density Curve}
]$$


üí¨ In simple terms:
> A box plot tells you how much and where,<br>

A **Histogram** is a bar-like representation that shows how data values are distributed across **intervals (bins)**.

Each bar‚Äôs height represents the **frequency** (or count) of observations within that bin.

$$[
\text{Frequency of a bin} = \text{Number of values falling within that range}
]$$

| Term | Meaning |
|:------|:-------------|
| **Bins** | The number of equal-width intervals used to group data |
| **KDE (Kernel Density Estimate)** | A smooth curve showing the probability density of the data |

# üåä KDE Plot (Kernel Density Estimate) in Seaborn

üìå **Objective:**  
Understand and visualize the **probability distribution** of a numerical variable using `sns.kdeplot()`.  
A KDE Plot shows where your data points are **most concentrated** by estimating the underlying continuous distribution.

---

## üß† What is a KDE Plot?

A **KDE (Kernel Density Estimate)** plot is a **smoothed version of a histogram**.  
Instead of grouping data into bars (bins), it draws a **continuous curve** representing the estimated probability density function (PDF).

$$[
\text{KDE}(x) = \frac{1}{n h} \sum_{i=1}^{n} K\left(\frac{x - x_i}{h}\right)
]$$

where:
- \( K \) = kernel function (usually Gaussian)
- \( h \) = bandwidth (controls smoothness)
- \( n \) = number of data points  

üí° *In simple terms:*  
> KDE plots show the **shape** of your data ‚Äî how it‚Äôs spread and where it peaks


````python
# --------------------------------------------------------
# üåä Basic KDE Plot using Seaborn
# --------------------------------------------------------

# üß† Step 1: Load dataset
tips = sns.load_dataset('tips')

# üß© Step 2: Create a KDE plot
plt.figure(figsize=(8, 5))
sns.kdeplot(
    data=tips['total_bill'],  # Data column to plot
    fill=True,                # Fill the area under the curve
    color='teal',             # Curve color
    linewidth=2,              # Curve thickness
    alpha=0.7                 # Transparency of fill
)

# üè∑Ô∏è Step 3: Add labels and title
plt.title('KDE Plot - Distribution of Total Bill', fontsize=14, fontweight='bold')
plt.xlabel('Total Bill ($)', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
````

‚úÖ Output Explanation
- The X-axis shows the variable‚Äôs range (total bill amounts).
- The Y-axis shows the density ‚Äî not the count, but relative frequency.
- The area under the curve = 1 (represents total probability).
- Peaks in the curve show where most data points lie (most common bill amounts).

---
## üß© Example 2: Comparing KDE Distributions by Gender


````python
# --------------------------------------------------------
# üåä Multiple KDE Curves using hue
# --------------------------------------------------------

plt.figure(figsize=(8, 5))
sns.kdeplot(
    data=tips, 
    x='total_bill',
    hue='sex',                # Separate curves for Male/Female
    fill=True,                # Fill under each curve
    common_norm=False,        # Keep densities separate (don‚Äôt normalize together)
    palette='Set2',           # Nice color palette
    alpha=0.6,                # Transparency for overlapping areas
    linewidth=2
)

plt.title('KDE Plot - Distribution of Total Bill by Gender', fontsize=14, fontweight='bold')
plt.xlabel('Total Bill ($)')
plt.ylabel('Density')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
````

‚úÖ Enhanced Output Insight
- Each curve represents the distribution for one gender.
- The filled area helps visualize overlap ‚Äî showing how spending patterns differ.
- You can clearly see if one group tends to have higher or lower bills.

| Tip                       | Description                                                        |

]$$

# üî• Heatmap in Seaborn

üìå **Objective:**  
Visualize the **correlation strength** between numerical variables using Seaborn‚Äôs `heatmap()`.  
A heatmap uses **color intensity** to represent how strongly variables are related ‚Äî  
making it easy to identify **positive, negative, or weak correlations**.

---

## üß† What is a Heatmap?

A **Heatmap** is a graphical representation of data where values are shown as **color shades**.  
In correlation analysis, it helps us visualize how one numerical variable changes in relation to another.

$$[
\text{Correlation Coefficient (r)} = \frac{\text{Cov}(X, Y)}{\sigma_X \sigma_Y}
]$$

Where:
- \( r \) ‚Üí ranges between **-1** and **+1**  
- \( r = +1 \) ‚Üí Perfect positive correlation  
- \( r = -1 \) ‚Üí Perfect negative correlation  
- \( r = 0 \) ‚Üí No correlation  

---

### üîπ Interpreting Colors

| Correlation (r) | Meaning | Color |
|:----------------:|:--------|:------|
| +1.0 | Strong positive relation | üî¥ Red |
|  0.0 | No relation | ‚ö™ Neutral (white/light) |
| -1.0 | Strong negative relation | üîµ Blue |

üí¨ **In simple words:**  
> A heatmap turns numbers into colors, so your brain spots relationships instantly! üéØ

---

## üß© Example 1: Basic Correlation Heatmap


````python
# --------------------------------------------------------
# üî• Correlation Heatmap using Seaborn
# --------------------------------------------------------

# üß† Step 1: Load dataset
tips = sns.load_dataset('tips')

# üßæ Step 2: Inspect data
display(tips.head())

# üßÆ Step 3: Compute correlation matrix
correlation = tips[['total_bill', 'tip', 'size']].corr()

# üß© Step 4: Plot heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(
    correlation,        # Matrix to visualize
    annot=True,         # Display correlation values
    cmap='coolwarm',    # Color palette (blue ‚Üí red)
    center=0,           # Center color map at 0
    linewidths=0.5,     # Add thin lines between cells
    fmt=".2f"           # Format annotation to 2 decimal places
)

# üè∑Ô∏è Step 5: Add title
plt.title('Correlation Heatmap of Numerical Features', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
````

‚úÖ Output Explanation
- Each cell shows the correlation coefficient (r) between two variables.
- Red ‚Üí strong positive relationship (values rise together).
- Blue ‚Üí strong negative relationship (one rises, the other falls).
- White/light shades ‚Üí weak or no relationship.

For example:
If `total_bill` and tip show a correlation near `+0.67`, it means larger bills usually mean larger tips üí∏

# üìà Regression & Trend Line Visualization (Seaborn)

üìå **Objective:**  
Fit and visualize linear (and simple nonlinear) relationships between two numerical variables using Seaborn's `regplot` and `lmplot`.

---

## üß† What is Regression (in this context)?

Regression fits a function (commonly a straight line) that best describes the relationship between an independent variable \(x\) and a dependent variable \(y\).

A simple linear regression model:

$$[
y \approx m x + b
]$$

where:
- \(m\) = slope (how much \(y\) changes for 1 unit of \(x\))  
- \(b\) = intercept

Seaborn's regression plotting functions visualize both the scatter of points and the fitted trend line (with optional confidence interval).


````python
# --------------------------------------------------------
# üìà Example 1: Simple Linear Regression with sns.regplot()
# --------------------------------------------------------

# Load sample dataset
tips = sns.load_dataset('tips')

# Create scatter + regression line (default: linear fit with 95% CI)
plt.figure(figsize=(8, 5))
sns.regplot(
    data=tips,
    x='total_bill',          # Independent variable
    y='tip',                 # Dependent variable
    scatter_kws={'s': 60, 'edgecolor': 'black', 'alpha': 0.7},  # scatter point style
    line_kws={'color': 'crimson', 'linewidth': 2},              # regression line style
    ci=95,                   # 95% confidence interval band for the regression line
    marker='o'               # marker style for scatter points
)

plt.title('regplot: Tip vs Total Bill (Linear Fit with 95% CI)', fontsize=14, fontweight='bold')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
````

### Explanation (regplot)
- `sns.regplot()` plots raw points and fits a regression model (by default linear).  
- `ci` shows an uncertainty band (confidence interval) around the fitted line ‚Äî helpful to assess how certain the estimate is.  
- `scatter_kws` and `line_kws` let you style points and the fitted line.

````python
# --------------------------------------------------------
# üìà Example 2: Linear Regression without Confidence Interval & robust fit
# --------------------------------------------------------

plt.figure(figsize=(8, 5))

# Ordinary linear fit without CI
sns.regplot(
    data=tips, x='total_bill', y='tip',
    ci=None,                    # Hide confidence interval
    scatter_kws={'s': 50, 'alpha': 0.6},
    line_kws={'color': 'navy', 'linewidth': 2, 'linestyle': '--'},
    label='OLS fit'
)

# Robust fit (low influence of outliers) using lowess (locally weighted scatterplot smoothing)
sns.regplot(
    data=tips, x='total_bill', y='tip',
    lowess=True,                # non-parametric locally-weighted regression (smooth curve)
    scatter=False,              # avoid plotting scatter again
    line_kws={'color': 'orange', 'linewidth': 2},
    label='LOWESS (smooth)'
)

plt.title('Linear vs LOWESS Fit (no CI)', fontsize=14, fontweight='bold')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
````

### Notes
- `lowess=True` draws a smooth non-parametric curve useful when the relationship is not strictly linear.  
- Use `ci=None` to hide the confidence band when you prefer a cleaner visual.


````python
# --------------------------------------------------------
# üìà Example 3: Use sns.lmplot() for faceted regression (fit per category)
# --------------------------------------------------------

# lmplot returns a FacetGrid ‚Äî great for comparing separate regression lines per group
sns.lmplot(
    data=tips,
    x='total_bill', y='tip',
    hue='sex',                 # separate regression per gender
    col='smoker',              # facet by smoker status (creates two panels)
    height=5, aspect=1,        # size of each facet
    scatter_kws={'s': 50, 'alpha': 0.6, 'edgecolor': 'black'},
    line_kws={'linewidth': 2}
)

# add a global title (suptitle)
plt.suptitle('lmplot: Regression by Sex and Smoker Status', fontsize=16, fontweight='bold', y=1.03)
plt.tight_layout()
plt.show()
````

### Why use `lmplot`?
- `lmplot` is handy when you want **separate fits per subgroup** (via `hue` or `col`/`row`) and automatic faceting.  
- It returns a `FacetGrid` so you can easily compare slopes/intercepts across categories.
